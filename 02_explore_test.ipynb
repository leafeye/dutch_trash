```python
#wss enige correcte values zijn 2 en 4

#new df with freq inzameling 2 OF 4
freqdf = df[(df.frequentie_van_inzameling_huishoudelijk_restafval == 2) | (df.frequentie_van_inzameling_huishoudelijk_restafval == 4)]

freqdf.head()
```

Lijkt erop dat men beter gaat scheiden als er meer frequent afval wordt opgehaald
Onverwachts vind ik!

Null hypothesis:
SCheidingspercentage verschilt niet significant tussen de twee verschillende inzamelfrequenties

Alternative:
Meer frequent is betere scheiding


```python
import researchpy as rp
from scipy import stats
```

```python
rp.summary_cont(freqdf.groupby("frequentie_van_inzameling_huishoudelijk_restafval")['scheidingspercentage_fijn_huishoudelijk_afval'])
```

```python
tweemaal = freqdf[(freqdf['frequentie_van_inzameling_huishoudelijk_restafval'] == 2)]
tweemaal.reset_index(inplace= True)

viermaal = freqdf[(freqdf['frequentie_van_inzameling_huishoudelijk_restafval'] == 4)]
viermaal.reset_index(inplace= True)
```

Check homogeneity of Variances met Levene's test

```python
stats.levene(tweemaal['scheidingspercentage_fijn_huishoudelijk_afval'], viermaal['scheidingspercentage_fijn_huishoudelijk_afval'])
```

Niet significant dus assumptie niet geschonden, doorgaan!

```python
descriptives, results = rp.ttest(viermaal['scheidingspercentage_fijn_huishoudelijk_afval'], tweemaal['scheidingspercentage_fijn_huishoudelijk_afval'])
```

```python
descriptives
```

```python
results
```

Het scheidingspercentage van fijn huishoudelijk afval ligt hoger bij gemeentes waar viermaal per maand wordt ingezameld (M=70.0, SD= 12.48) dan bij gemeentes met tweemaal per maand (M = 56.33, SD = 12.63). Er is een signifivant verschil tussen het scheidingspercentage van de tweemaal en viermaal maandelijkse inzalinggemeentes (t(292)= 8.499, p < 0.0001, r = 0.445)

Heel significant verschil!
