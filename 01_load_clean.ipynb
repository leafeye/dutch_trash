Create column for provincie and gemeente by splitting the string in Provincies column

```python
#store in temporary df splitdf
splitdf = df.Provincies.str.split(': ', expand=True)  #with expand = true the split will be expanded into separate columns

splitdf.head(3)
```

```python
#add the provincie and gemeente column to df
df['provincie'] = splitdf.iloc[:,0]
df['gemeente'] = splitdf.iloc[:,1]
```

```python
#drop Probincies column
df = df.drop(columns='Provincies')

df.head(2)
```
