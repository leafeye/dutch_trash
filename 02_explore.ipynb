```python
df = pd.read_csv('data/gemeenten_trash.csv')
df.head()
```

```python
df.describe()
```

```python
#welke provincie produceert gemiddeld het meeste huishoudelijke afval?

prov_tot = df.loc[ : , ['provincie', 'totaal_aangeboden_huishoudelijk_afval']]
prov_tot.groupby('provincie').mean()
```

```python
plt.figure(figsize=(8,6))

sns.set(style="whitegrid")

ax = sns.barplot(x="provincie", y="totaal_aangeboden_huishoudelijk_afval", data=prov_tot)
plt.xticks(rotation=45)
```

```python
gem_tot = df.loc[:,['provincie', 'gemeente', 'totaal_aangeboden_huishoudelijk_afval' ]]

```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.set_style("white")

#plt.suptitle('Album Scrobbles', x=0.2)

sns.boxplot(y=gem_tot["provincie"],x=gem_tot["totaal_aangeboden_huishoudelijk_afval"])

plt.ylabel('')
plt.xlabel('totaal_aangeboden_huishoudelijk_afval')

#sns.stripplot(x=ween["scrobbles"], y=ween["album"], color="orange", jitter=0.2, size=3.5) #points to show the values
sns.despine(left=True, bottom = True)
```

```python
#friesland has big outlier!
#gemeentes with lowest trash are in noordbrabant and overijssel
```

```python
#gemeentes met laagst aangeboden afval
gem_tot.sort_values('totaal_aangeboden_huishoudelijk_afval').head()
```

```python
#gemeentes met hoogst aangeboden afval
gem_tot.sort_values('totaal_aangeboden_huishoudelijk_afval', ascending=False).head(10)
```

```python
#seems like low population gemeentes has higher total per persoon
```
